```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020='athletic_sales_2020.csv'
sales_2021='athletic_sales_2021.csv'

df_sales_2020=pd.read_csv(sales_2020)
df_sales_2021=pd.read_csv(sales_2021)

```

```python
# Display the 2020 sales DataFrame
df_sales_2020.head()
```

```python
# Display the 2021 sales DataFrame
display(df_sales_2021.head())
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_sales_2020.head()
```

```python
# Check the 2021 sales data types.
df_sales_2021.head()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combine_df=pd.concat([df_sales_2020, df_sales_2021], ignore_index=True)
combine_df.reset_index(drop=True, inplace=True)
display(combine_df.head())
```

```python
# Check if any values are null.
#null_values=combine_df.isnull().any()
#display(null_values[null_values==True])

null_values=combine_df.isnull().sum()
display(null_values)
```

```python
# Check the data type of each column
display(combine_df.dtypes)
```

```python
# Convert the "invoice_date" to a datetime datatype
combine_df['invoice_date']=pd.to_datetime(combine_df['invoice_date'])
display(combine_df.dtypes)
```

```python
# Confirm that the "invoice_date" data type has been changed.
display(combine_df.dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
total_products_sold_df = combine_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
# Rename the sum to "Total_Products_Sold".
total_products_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.
top_5_results = total_products_sold_df.sort_values(by='Total_Products_Sold', ascending=False).head(5)  
display(top_5_results)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold_df = pd.pivot_table(combine_df,index=['region', 'state', 'city'],values='units_sold',aggfunc='sum').reset_index()
# Rename the "units_sold" column to "Total_Products_Sold"
total_products_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
# Show the top 5 results.
display(top_5_results)

```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_df = combine_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
# Rename the "total_sales" column to "Total Sales"
total_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
top_5_regions = total_sales_df.sort_values(by='Total Sales', ascending=False).head(5)
display(top_5_regions)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_df = pd.pivot_table(combine_df, index=['region', 'state', 'city'], values='total_sales', aggfunc='sum').reset_index()
# Optional: Rename the "total_sales" column to "Total Sales"
total_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
top_5_regions = total_sales_df.sort_values(by='Total Sales', ascending=False).head(5)
display(top_5_regions)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sales_df = combine_df.groupby(['retailer', 'region', 'state', 'city']).agg({'total_sales': 'sum'}).reset_index()
# Rename the "total_sales" column to "Total Sales"
retailer_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
top_5_retail_df = retailer_sales_df.sort_values(by='Total Sales', ascending=False).head(5)
display(top_5_retail_df)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sales_pdf = pd.pivot_table(combine_df, index=['retailer', 'region', 'state', 'city'], values='total_sales', aggfunc='sum').reset_index()
# Optional: Rename the "total_sales" column to "Total Sales"
retailer_sales_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the top 5 results.
top_5_retail_df = retailer_sales_df.sort_values(by='Total Sales', ascending=False).head(5)
display(top_5_retail_df)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_shoes_df = combine_df[combine_df['product'] == "Women's Athletic Footwear"]
display(womens_shoes_df)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
total_womens_shoes_df = womens_shoes_df.groupby(['retailer', 'region', 'state', 'city']).agg({'units_sold': 'sum'}).reset_index()
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_womens_shoes_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
top_5_womens_shoes_df = total_womens_shoes_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
display(top_5_womens_shoes_df)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
total_womens_shoes_df = pd.pivot_table(womens_shoes_df, index=['retailer', 'region', 'state', 'city'], values='units_sold', aggfunc='sum').reset_index()
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
total_womens_shoes_df.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
top_5_womens_shoes_df = total_womens_shoes_df.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).head(5)
display(top_5_womens_shoes_df)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
total_womens_shoes_df = pd.pivot_table(womens_shoes_df, index='invoice_date', values='total_sales', aggfunc='sum')
# Optional: Rename the "total_sales" column to "Total Sales"
total_womens_shoes_df.rename(columns={'total_sales': 'Total Sales'}, inplace=True)
# Show the table.
display(total_womens_shoes_df)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales_df = total_womens_shoes_df.resample('D').sum()
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales_sorted_df = daily_sales_df.sort_values(by='Total Sales', ascending=False)
display(daily_sales_sorted_df)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_bin_df = total_womens_shoes_df.resample('W').sum()
# Sort the resampled pivot table in descending order on "Total Sales".
weekly_bin_total_df = weekly_bin_df.sort_values(by='Total Sales', ascending=False)
display(weekly_bin_total_df)
```
